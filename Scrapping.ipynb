```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.panditfootball.com/"
response = requests.get("https://www.panditfootball.com/")

data = BeautifulSoup(response.text, "html.parser")
response.content
```

```python
cards_data = data.find_all('article', attrs={'class':'news-block small-block'})

for card in cards_data:
    nama_manga = card.find('a'), attrs={'class':'container['class':'row']'}).text
    print(nama_manga)
```
